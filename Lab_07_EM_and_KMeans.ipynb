```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
import sklearn.metrics as sm
from sklearn.cluster import KMeans
iris = datasets.load_iris()
print("IRIS DATA: \n",iris.data)
X=pd.DataFrame(iris.data)
print(X)
X.columns=['s_l','s_w','p_l','p_w']
y=pd.DataFrame(iris.target)
y.columns=['target']
print(y)
```

```python
plt.figure(figsize=(14,7))
colormap = np.array(["red","lime","black"])
plt.subplot(1,2,1)
plt.scatter(X.s_l,X.s_w,c=colormap[y.target],s=40)
plt.title("Sepal")
plt.subplot(1,2,2)
plt.scatter(X.p_l,X.p_w,c=colormap[y.target],s=40)
plt.title("Petal")
```

```python
model = KMeans(n_clusters=3,random_state=100)
model.fit(X)
y_pred = model.labels_
print(y_pred)
```

```python
plt.subplot(1,2,1)
plt.scatter(X.p_l,X.p_w,c=colormap[y.target],s=40)
plt.title("Real classification")
plt.subplot(1,2,2)
plt.scatter(X.p_l,X.p_w,c=colormap[y_pred],s=40)
plt.title("KMeans classification")
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture( n_components=3,covariance_type='full',random_state=100)
gmm.fit(X)
y_clusters_gmm = gmm.predict(X)
y_clusters_gmm
```

```python
plt.figure(figsize=(16,7))
plt.subplot(1,3,1)
plt.scatter(X.p_l,X.p_w,c=colormap[y.target],s=40)
plt.title("Real classification")
plt.subplot(1,3,2)
plt.scatter(X.p_l,X.p_w,c=colormap[y_pred],s=40)
plt.title("KMeans classification")
plt.subplot(1,3,3)
plt.scatter(X.p_l,X.p_w,c=colormap[y_clusters_gmm],s=40)
plt.title("GMM classification")
```

```python
print("KMeans Prediction:")
print("Accuracy: \n",sm.accuracy_score(y,y_pred))
print("Confusion_matrix: \n",sm.confusion_matrix(y,y_pred))
print("GMM Prediction:")
print("Accuracy: \n",sm.accuracy_score(y,y_clusters_gmm))
print("Confusion_matrix: \n",sm.confusion_matrix(y,y_clusters_gmm))

```
