**Implementing AO Star algorithm**

```python
class Graph:
  def __init__(self,graph,heuristicNodeValueList,startNode):
    self.graph = graph
    self.H = heuristicNodeValueList
    self.start = startNode
    self.parent = {}
    self.status = {}
    self.solutiongraph = {}
  def  applyAOStar(self):
    self.aostar(self.start, False)
  def  getNeighbors(self,v):
    return self.graph.get(v,'')
  def getHeuristicNodeValue(self,n):
    return self.H.get(n,0)
  def setHeuristicNodevalue(self,n,val):
     self.H[n]=val
  def getStatus(self,v):
    return self.status.get(v,0)
  def setStatus(self,v,val):
    self.status[v]=val
  def printSolution(self):
    print("Traverse graph from start ",self.start)
    print("------------------------------------")
    print("Graph solution",self.solutiongraph)
    print("------------------------------------")
  def compute_cost(self,v):
    min_cost=0
    child_dict={}
    child_dict[min_cost]=[]
    flag = True
    for nodeInfo in self.getNeighbors(v):
      cost = 0
      node_list = []
      for c,w in nodeInfo:
        cost= cost + self.getHeuristicNodeValue(c)+w
        node_list.append(c)
      
      if flag == True:
        min_cost = cost
        child_dict[min_cost] = node_list
        flag = False
      else:
         if min_cost > cost:
           min_cost = cost
           child_dict[min_cost] = node_list
    return min_cost,child_dict[min_cost]

  def aostar(self,v,backtracking):
    print("HEURISTIC VALUE",self.H)
    print("GRAPH SOLUTION ",self.solutiongraph)
    print("-----------------------------------")
    print("PROCESSING NODE",v)
    if self.getStatus(v)>=0:
      min_cost,child_list = compute_cost(self,v)
      self.setHeuristicNodevalue(v,min_cost)
      self.setStatus(v,len(child_list))
      solved = True
      for childnode in child_list:
        self.parent[childnode]=v
        if self.getStatus(childnode)!=-1:
          solved = solved & False
      if solved == True:
        self.setStatus(v,-1)
        self.solutiongraph[v] = child_list
      if v!=self.start:
        self.aostar(self.parent[v],True)
      if backtracking == False:
        for childnode in child_list:
          self.setStatus(childnode,0)
          self.aostar(childnode,False)

```

```python
h1 = {'A': 1, 'B': 6, 'C': 2, 'D': 12, 'E': 2, 'F': 1, 'G': 5, 'H': 7, 'I': 7, 'J': 1, 'T': 3}
graph1 = {
    'A': [[('B', 1), ('C', 1)], [('D', 1)]],
    'B': [[('G', 1)], [('H', 1)]],
    'C': [[('J', 1)]],
    'D': [[('E', 1), ('F', 1)]],
    'G': [[('I', 1)]]   
}
G1= Graph(graph1, h1, 'A')
G1.applyAOStar() 
G1.printSolution()
```
